```python
import pandas as pd
```

```python
filepath = "DATA/RAW/table2.csv"
df = pd.read_csv(filepath)

# Columns that are actual parameters
exclude_cols = ['bsdb', 'ref_idsys', 'id', 'catalogue', 'filename', 'line']
param_cols = [col for col in df.columns if col not in exclude_cols]

param_catalogue_map = {}

for param in param_cols:
    subset = df[df[param].notna()]
    catalogues = sorted(subset['catalogue'].dropna().unique().tolist())
    param_catalogue_map[param] = catalogues

print("LIST OF PARAMETERS AND SOURCES")
for param, cats in param_catalogue_map.items():
    print(f"{param} ({len(cats)}):", {', '.join(cats)})
```
